# We are going to calculate the following integration:

$\int_0^{\infty} x^2\frac{sin(Qx)}{Qx} dx$ (*1*)

## First of all, we have:

```
Integrate[Exp[-x*t],{t,0,Infinity}] (*2*)
```

Therefore, we can replace the 1/x in the original integrand (*1*) with the integration above (*2*).

Then the integration over x and t can be switched in order and we therefore have the following integration as the result.

```
Integrate[Integrate[x^2*Sin[Q*x]*Exp[-x*t],{x,0,Infinity}],{t,0,Infinity}] (*3*)
```

In practice we have Q as a real number for scattering experiment (i.e. Im[Q]=0). Therefore, when Q>0 (i.e. Re[Q]>0, which is the first condition as given above (*3*)), the integration we want to find at the very beginning gives 0. When Q=0, the conditional expression given above (*3*) can be simplified, since we have (going back to the original integral (*1*)):

$\lim_{Qx\rightarrow 0}\frac{sin(Qx)}{Qx}=1$

In such a situation, the original integral (*1*) simply goes to infinity. However, in practice, since we don't have material with infinite size, the integral will give some positive value, which indicates that at Q=0, we have a strong scattering feature, which, however, is inaccessible experimentally (since usually, this Q=0 scattering is mixed together with the beam that actually does not interact at all with the measured sample).
