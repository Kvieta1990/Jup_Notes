```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

Define the GP squared exponential kernel $\kappa(\mathbf{x}_1, \mathbf{x}_2)= e^{-\mathcal{l}||\mathbf{x}_1 - \mathbf{x}_2||}$ as follows (where $\mathcal{l}$ is a pre-defined or to-be-fitted parameter):

```python
def kernel(a, b):
    sqdist = np.sum(a**2, 1).reshape(-1,1) + np.sum(b**2, 1) - 2*np.dot(a, b.T)
    return np.exp(-.5 * sqdist)
```

```python
n = 50 # Number of test points.
X_test = np.linspace(-5, 5, n).reshape(-1 ,1) # Test points, reshaped to a 50*1 array.
K_ = kernel(X_test, X_test) # Kernel at test points themselves.
```

Here the formulation $\mathbf{f} = \mathbf{L}\cdot N(0, 1)$ will be used to generate multivariate Gaussian values. In thise case, we have 50 variables and we want to generate them 10 times, i. e. we have 10 sets of 50 multivariate Gaussian values in total.

```python
L = np.linalg.cholesky(K_ + 1e-6 * np.eye(n)) # Cholesky decomposition
f_prior = np.dot(L, np.random.normal(size=(n, 10)))
```

```python
plt.plot(X_test, f_prior);
```
